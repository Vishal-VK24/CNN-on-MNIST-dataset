```python
import tensorflow as tf
import numpy as np
from tensorflow import keras
import matplotlib.pyplot as plt
```

```python
(x_train,y_train),(x_test,y_test)= keras.datasets.mnist.load_data()
```

```python
print(len(x_train))
print(x_train[0].shape)
print(x_train[0])
```

```python
plt.matshow(x_train[0])
```

```python
x_train_flat=x_train.reshape(len(x_train),28*28)
x_test_flat=x_test.reshape(len(x_test),28*28)
print(x_train_flat.shape)
print(x_test_flat.shape)
x_train_flat= x_train_flat /255
x_test_flat= x_test_flat /255
```

```python
model=keras.models.Sequential([
            keras.layers.Dense(60, input_dim=28*28, activation='relu'),
            keras.layers.Dense(24, activation='relu'),
            keras.layers.Dense(10, activation='sigmoid')

])
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
model.fit(x_train_flat,y_train,epochs=6)
```

```python
predicted= model.predict(x_test_flat)
predicted=[np.argmax(i) for i in predicted]
print(predicted)
```

```python
print(tf.math.confusion_matrix(y_test,predicted))
```

```python
model.evaluate(x_test_flat,y_test)
```
